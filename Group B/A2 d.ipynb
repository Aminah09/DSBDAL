```python
cp_counts = df1['cp'].value_counts()

plt.pie(cp_counts, labels=cp_counts.index, autopct='%1.1f%%')
plt.title('Distribution of cp Variable')
plt.axis('equal')
plt.show()
```

```python
age_values = df1['age']

plt.fill_between(range(len(age_values)), age_values)
plt.title('Area Plot for Age')
plt.xlabel('Data Points')
plt.ylabel('Age')
plt.show()

```
